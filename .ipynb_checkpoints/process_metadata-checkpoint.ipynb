```python
# Hàm xử lý: chỉ cắt ký tự cuối nếu là '0Y', còn lại cắt ký tự đầu và cuối
def process_age(age):
    if age == '0Y':
        return age[:-1]  # Chỉ cắt ký tự cuối (chữ cái 'Y')
    else:
        return age[1:-1]  # Loại bỏ ký tự đầu và cuối

# Áp dụng hàm cho cột 'age'
metadata["age"] = metadata["age"].apply(lambda x: process_age(x))

# Kiểm tra kết quả
metadata["age"]
```

```python
import pandas as pd

def balance_dataset(file_path, output_path, birads_target='BI-RADS 2', ratio=2):
    # Đọc file CSV
    df = pd.read_csv(file_path)
    
    # Số lượng hàng mong muốn cho BI-RADS 1
    birads_2_count = len(df[df['breast_birads'] == birads_target])
    birads_1_target_count = birads_2_count * ratio

    # Lọc các hàng không phải BI-RADS 1
    non_birads_1 = df[df['breast_birads'] != 'BI-RADS 1']
    
    # Lấy ngẫu nhiên một số hàng từ BI-RADS 1
    birads_1 = df[df['breast_birads'] == 'BI-RADS 1']
    sampled_birads_1 = birads_1.sample(n=birads_1_target_count, random_state=42)

    # Kết hợp lại
    balanced_df = pd.concat([non_birads_1, sampled_birads_1], axis=0).reset_index(drop=True)

    # Ghi lại vào file mới
    balanced_df.to_csv(output_path, index=False)

# Đường dẫn dữ liệu gốc và mới
train_label = '/media/mountHDD3/data_storage/biomedical_data/Dataset/VinDr-Mammo/Processed/train_df.csv'
val_label = '/media/mountHDD3/data_storage/biomedical_data/Dataset/VinDr-Mammo/Processed/val_df.csv'
test_label = '/media/mountHDD3/data_storage/biomedical_data/Dataset/VinDr-Mammo/Processed/test_df.csv'

balanced_train_label = '/media/mountHDD3/data_storage/biomedical_data/Dataset/VinDr-Mammo/Processed/balanced_train_df.csv'
balanced_val_label = '/media/mountHDD3/data_storage/biomedical_data/Dataset/VinDr-Mammo/Processed/balanced_val_df.csv'
balanced_test_label = '/media/mountHDD3/data_storage/biomedical_data/Dataset/VinDr-Mammo/Processed/balanced_test_df.csv'

# Tạo các tập dữ liệu cân bằng
train_2 = balance_dataset(train_label, balanced_train_label)
val_2 = balance_dataset(val_label, balanced_val_label)
test_2 = balance_dataset(test_label, balanced_test_label)
```

```python
import pandas as pd

# Giả sử dataframe của bạn được lưu trong biến `df`
# Tạo cột đếm số lần xuất hiện của mỗi giá trị trong cột 'image_id'
counts = metadata['image_id'].value_counts()

# Lấy các giá trị 'image_id' xuất hiện không quá 2 lần
valid_image_ids = counts[counts < 2].index

# Lọc dataframe để chỉ giữ lại các dòng có 'image_id' hợp lệ
filtered_df = metadata[metadata['image_id'].isin(valid_image_ids)]

filtered_df
```

```python
# Định nghĩa mapping để thay thế các giá trị
mapping = {
    'BI-RADS 1': 'normal',
    'BI-RADS 2': 'benign',
    'BI-RADS 3': 'benign',
    'BI-RADS 4': 'malignant',
    'BI-RADS 5': 'malignant'
}

# Áp dụng thay thế trên từng DataFrame
filtered_df['breast_birads'] = filtered_df['breast_birads'].replace(mapping)
 

 
print(filtered_df['breast_birads'].value_counts())
```

```python
filtered_df['breast_birads'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
train_df, temp_df = train_test_split(
    filtered_df,
    test_size=0.2,
    stratify=filtered_df['breast_birads'],  # Đảm bảo phân bổ theo breast_birads
    random_state=42
)

# Bước 2: Chia dữ liệu test+val (20%) thành test (10%) và val (10%)
test_df, val_df = train_test_split(
    temp_df,
    test_size=0.5,  # 50% của 20% là 10% tổng dữ liệu
    stratify=temp_df['breast_birads'],  # Đảm bảo phân bổ theo breast_birads
    random_state=42
)
```
